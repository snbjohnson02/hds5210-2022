```python
months =  ['April', 'August', 'February', 'January', 'July', 'June', 'March', 'May','September', 'October','November']
monthlydata= pd.DataFrame()
for m in months:
    filename = m + ' 2022.csv'
    monthfile=pd.read_csv(filename)
    monthlydata= monthlydata.append(monthfile)
```

```python
monthlyreport=monthlydata.groupby(['Day of the Week'])['No Shows', '# of Patients Scheduled'].sum()
monthlyreport['No Show Rate'] = monthlyreport['No Shows']/monthlyreport['# of Patients Scheduled']
monthlyreport
# Look at no shows by Day of the week and total number of patients.  Sum of no shows and # of patients scheduled
# No show rate monthly report
```

```python
monthlydata['Date'] = monthlydata['Date'].astype('datetime64[ns]')
monthlydata['Month'] =monthlydata['Date'].dt.month
monthlydatareport = pd.DataFrame(monthlydata.groupby('Month')['# of Patients Scheduled', 'No Shows',
       'Total Patient\'s Completed'].sum())
monthlydatareport['Percent Completed'] = monthlydatareport['Total Patient\'s Completed']/ monthlydatareport['# of Patients Scheduled']
```
