**FINAL RESULT**

For a final result, I'd like you to produce a data frame named **summary** that contains:
* One ROW for each userName.  The userName can either be the row index or a column in the data frame.
* One COLUMN for the "BEFORE" week and one COLUMN for the "DURING" week.  Name those columns 'Before' and 'During'.
* One more calculated column that shows the percent increase from BEFORE to DURING.  Name that column 'pctChange'.


```python
import pandas as pd
meetings = pd.read_csv('/data/meetings.csv')
```

```python
meetings.loc[(meetings.startTime.astype('datetime64[ns]').dt.day >=17) &
(meetings.startTime.astype('datetime64[ns]').dt.day <=28) &
(meetings.startTime.astype('datetime64[ns]').dt.month ==2), 'Period'] = 'Before'

meetings.loc[(meetings.startTime.astype('datetime64[ns]').dt.day >=16) &
(meetings.startTime.astype('datetime64[ns]').dt.day <=27) &
(meetings.startTime.astype('datetime64[ns]').dt.month ==3), 'Period'] = 'During'
meetings.Period.value_counts(dropna=False)
```

```python
summary = pd.crosstab(meetings['userName'], meetings['Period'])
summary['pctchange'] = (summary['During']-summary['Before']) /summary['Before']
summary
```
